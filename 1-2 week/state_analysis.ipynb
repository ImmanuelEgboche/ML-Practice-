```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('every_state.csv')

for state in df['RegionName'].unique():
    state_data = df[df['RegionName'] == state].copy()  # Create a copy here
    state_data['Date'] = pd.to_datetime(state_data['Year'].astype(str) + '-' + state_data['Month'].astype(str))
    
    plt.figure(figsize=(10, 6))
    plt.plot(state_data['Date'], state_data['HousePrice'])
    plt.xlabel('Date')
    plt.ylabel('HousePrice')
    plt.title(f'House Prices Over Time in {state}')
    plt.xticks(rotation=45)
    plt.show()
```
